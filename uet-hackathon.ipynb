```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split as tts
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
```

```python
work_train = pd.read_csv('../input/uet-hackathon-2022-data-science/work_train.csv')
info_train = pd.read_csv('../input/uet-hackathon-2022-data-science/info_train.csv')
work_test = pd.read_csv('../input/uet-hackathon-2022-data-science/work_test.csv')
info_test = pd.read_csv('../input/uet-hackathon-2022-data-science/info_test.csv')
res_train = pd.read_csv('../input/uet-hackathon-2022-data-science/label_train.csv')
test_output = pd.read_csv('../input/uet-hackathon-2022-data-science/label_test.csv')
```

```python
train = pd.merge(info_train, work_train, on='id_bh', how="inner")
```

```python
test = pd.merge(info_test, work_test, on='id_bh', how="inner")
```

```python
train = pd.merge(train, res_train, on='id_bh', how="inner")
```

```python
train = train.dropna()
```

```python
x_train = train.drop(['label','gender', 'address_x', 'id_office', 'job/role', 'address_y'],axis=1)
y_train = train['label']
x_test = test.drop(['gender', 'address_x', 'id_office', 'job/role', 'address_y'],axis=1)
```

```python
x_train.info()
```

```python
categorical_var = [col for col in x_train.columns if x_train[col].dtypes == 'O']
categorical_var
```

```python
x_test = x_test.drop_duplicates(subset=['id_bh'], keep='first')
```

```python
x_test.info()
```

```python
x_test.info()
```

```python
x_test['bithYear'].fillna('-1', inplace=True)
x_test['id_bh'].fillna('-1', inplace=True)
x_test['id'].fillna('-1', inplace=True)
x_test['id_management'].fillna('-1', inplace=True)
x_test['company_type'].fillna('-1', inplace=True)
x_test['from_date'].fillna('-1', inplace=True)
x_test['to_date'].fillna('-1', inplace=True)
x_test['employee_lv'].fillna('-1', inplace=True)
```

```python
numerical_var = [col for col in x_train.columns if x_train[col].dtypes != 'O']
numerical_var
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train,y_train)
```

```python
prediction = model.predict(x_test)
len(prediction)
```
